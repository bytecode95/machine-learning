Multiple Linear Regression Model
Multiple Linear Regression is an extension of Simple Linear Regression where we use multiple independent variables (features) to predict a single dependent variable.
Experience → Salary  ----> Simple Linear Regression Model

Experience
Education
Age
Skills
Location → Salary  ----> Multiple Linear Regression Model


Important Assumptions of Multiple Linear Regression

1. Linearity - Relationship should be linear.
2. No Multicollinearity - Features should not be highly correlated.
3. Homoscedasticity - Error variance should remain constant.
4. Normal Distribution of Residuals - Errors should follow normal distribution.
5. Independence of Observations - One observation should not depend on another.

How are categorical variables handled in Multiple Linear Regression?

Multiple Linear Regression cannot directly process categorical data. Categorical variables must be converted into numerical features using One Hot Encoding. To avoid the Dummy Variable Trap and multicollinearity, one dummy variable is dropped using OneHotEncoder(drop='first'). This allows the regression model to learn the impact of each category relative to a baseline category.

Model Evaluation Metrics
1. R² Score - Measures goodness of fit.
2. MAE - Mean Absolute Error
3. MSE - Mean Squared Error
4. RMSE - Root Mean Squared Error

Encoding Categorical Variables

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
print(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.2, random_state = 0)
```

Dummy Variable Trap
What is it?

Dummy Variable Trap occurs when one dummy variable can be perfectly predicted from other dummy variables.

This creates: Multicollinearity
which violates one of the assumptions of Multiple Linear Regression.

Suppose State column contains:
    California
    Florida
    New York

After OneHotEncoding 
| California | Florida | New York |
| ---------- | ------- | -------- |
| 1          | 0       | 0        |
| 0          | 1       | 0        |
| 0          | 0       | 1        |

Notice:
California = 1 - Florida - New York
California + Florida + New York = 1

Why is this a Problem?

The regression model cannot uniquely determine coefficients.
Profit =
b0 +
b1*California +
b2*Florida +
b3*NewYork

California + Florida + New York = 1

How to Fix It?

Drop one dummy column.
Instead of:
| California | Florida | New York |
| ---------- | ------- | -------- |
| 1          | 0       | 0        |
| 0          | 1       | 0        |
| 0          | 0       | 1        |

Use:
| Florida | New York |              |
| ------- | -------- | ------------ |
| 0       | 0        | ← California |
| 1       | 0        | ← Florida    |
| 0       | 1        | ← New York   |

California becomes the reference category.

The expected answer is:

Yes, Dummy Variable Trap should be avoided in Multiple Linear Regression by dropping one dummy variable from each categorical feature.

Training the Multiple Linear Regression Model on Training Set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

Predicting the Test Results

```python
y_pred = regressor.predict(x_test)
np.set_printoptions(precision=2)
print(y_pred)
```

```python
print(
    np.concatenate(
        (
            y_pred.reshape(len(y_pred), 1),
            y_test.to_numpy().reshape(len(y_test), 1)
        ),
        axis=1
    )
)
```
