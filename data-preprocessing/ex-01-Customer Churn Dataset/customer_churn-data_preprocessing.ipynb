```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('customer_churn.csv')
print(dataset)
print(dataset.shape)
print(dataset.describe)
```

```python
x =  dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
print(x)
print(y)
```

```python
import pandas as pd

x_df = pd.DataFrame(
    x,
    columns=[
        'customer_id',
        'age',
        'gender',
        'monthly_charges',
        'tenure_months',
        'contract_type'
    ]
)

print(x_df.head())
```

```python
print(dataset.isnull().sum())
```

```python
from sklearn.impute import SimpleImputer

gender_imputer = SimpleImputer(strategy='most_frequent')
x[:, 2:3] = gender_imputer.fit_transform(x[:, 2:3])

monthly_imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
x[:, 3:4] = monthly_imputer.fit_transform(x[:, 3:4])

contract_imputer = SimpleImputer(strategy='most_frequent')
x[:, 5:6] = contract_imputer.fit_transform(x[:, 5:6])
print(x)
```

```python
print(pd.isnull(x).sum())
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(
    transformers=[
        ('encoder', OneHotEncoder(), [2, 5])
    ],
    remainder='passthrough'
)

x = ct.fit_transform(x)

print(x)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
print('x_train', x_train)
print('x_test', x_test)
print('y_train',  y_train)
print('y_test', y_test)
```
