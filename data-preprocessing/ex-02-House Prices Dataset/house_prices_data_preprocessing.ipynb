```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('house_prices.csv')
print(dataset)
print(dataset.shape)
```

```python
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
print(x)
print(y)
```

```python
print(dataset.isnull().sum())
```

```python
from sklearn.impute import SimpleImputer
sqft_impute = SimpleImputer(missing_values=np.nan, strategy='mean')
sqft_impute.fit(x[:, 1:2])
x[:, 1:2] = sqft_impute.transform(x[:, 1:2])

location_impute = SimpleImputer(strategy='most_frequent')
location_impute.fit(x[:, 4:5])
x[:, 4:5] = location_impute.transform(x[:, 4:5])
print(x)
```

```python
print(pd.isnull(x).sum())
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(
    transformers=[
        ('encoder', OneHotEncoder(), [4])
    ],
    remainder='passthrough'
)
x = ct.fit_transform(x)
print(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
print('x_train', x_train)
print('x_test', x_test)
print('y_train',  y_train)
print('y_test', y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train[:, 3:] = sc.fit_transform(x_train[:, 3:])
x_test[:, 3:] = sc.transform(x_test[:, 3:])
print(x_train)
```
