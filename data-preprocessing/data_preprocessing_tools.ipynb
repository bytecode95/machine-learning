Taking Care of Missing Data
there is two ways of handling this
1. removing the missing data, but this will be okay if the percentage of the missing values of dataset is low
2. replacing the missing values with the average of that column

scikit-learn is python library which contain lot of tools for preprocessing data
SimpleImputer is a preprocessing utility used to handle missing values. With strategy='mean', it computes the average of each selected numerical column during fit() and replaces missing values with those averages during transform()

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(x[:, 1:3])
x[:, 1:3] = imputer.transform(x[:, 1:3])
print(x)
```

Encoding Categorical Data
machine learning models work with numbers, when dataset contains string input or output we need to convert them from string to numbers
we need carefully do this otherwise misinterpretation can be occur.
one-hot encoding - consists of creating binary vectors of each category
ex: France [100]
    Spain [010]
    Germany [001]
    there is no numerical order among these three countries
    otherwise if we take 1,2,3 it means there is a order

Encoding Independent Variables - use two classes from scikit-learn library
1. column transformer class from compose module
2. one hot encoded class from pre-processing module
ColumnTransformer applies preprocessing to specific columns. Here, OneHotEncoder is applied to the first column to convert categorical values into binary indicator features, while the remaining columns are passed through unchanged. The result is a fully numerical feature matrix suitable for machine learning algorithms.


```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
print(x)
```

Encoding dependent variable

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
print(y)
```

Splitting Dataset into training and test sets

Feature Scaling should apply after splitting dataset into training and test
feature scaling simply consists all variables and features to make sure all values take in the same scale 
Test set need to be a brand new set to evaluate the model
80% training set
20% test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
print('x_train', x_train)
print('x_test', x_test)
print('y_train',  y_train)
print('y_test', y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train[:, 3:] = sc.fit_transform(x_train[:, 3:])
x_test[:, 3:] = sc.transform(x_test[:, 3:])
print(x_train)
```
